## Section 3 — Build a Tidy DataFrame from Likes

```python
# converting data to DataFrame and assign it to a new variable using pandas
df = pd.DataFrame(dict)
df.head()
```

```python
dict.keys()
```

```python
df_tidy = pd.DataFrame(dict['likes_media_likes'])
df_tidy.head()
```

```python
# grabbing first element from list to see how data is structured
df_tidy.iloc[0]
```

## Section 4 - Create a Timestamp Column

```python
# isolating timestamp value 
df_tidy['string_list_data'].iloc[0]
```

```python
# returns object w/o brackets
df_tidy['string_list_data'].iloc[0][0]
```

```python
df_tidy['string_list_data'].iloc[0][0]['timestamp']
```

```python
df_tidy.drop(columns = ['string_list_data'])
```

```python
# how to create a new column (extracting timestamp from table)

df_tidy['Time of Like'] = [ i [0]['timestamp'] for i in df_tidy['string_list_data']] 

df_tidy.head()
```

```python
# converting numbers into readable date
from datetime import datetime
df_tidy['Date of Like'] = [datetime.fromtimestamp(i[0]['timestamp'])for i in df_tidy['string_list_data']]
```

```python
df_tidy.head()
```

```python
df_tidy['href'] = [ i [0]['href'] for i in df_tidy['string_list_data']] 
df_tidy.head()
```

## Section 5 — Summarize Likes by Account

```python
# grouping likes based on instagram handle or user
df_tidy.groupby('title').count().sort_values('Time of Like', ascending = False)
```
